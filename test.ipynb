```python
class Info(dspy.Signature):
    "Extract information from text passage"

    text: str = dspy.InputField(desc="Text to extract information from")
    name: str = dspy.OutputField(desc="Name extracted from text")
    age: int = dspy.OutputField(desc="Age extracted from text")
    role: str = dspy.OutputField(desc="Role extracted from text")
```

```python
predict = dspy.Predict(Info)
```

```python
out = predict(text="I'm Esai, 21 AI Dev")
```
